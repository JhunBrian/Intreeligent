```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
root = "final_logs/instance_segmentation/version_0/"

ckpt_root = os.path.join(root, "checkpoints")
ckpt_dict = {
    'epoch': [],
    'ckpt_path': [],
    'inference_path': []
}

for fldr in os.listdir(ckpt_root):
    fldr_path = os.path.join(ckpt_root, fldr)
    if os.listdir(fldr_path):
        e = fldr.split('=')[1].split('-')[0]
        ckpt_dict['epoch'].append(e)
        for file in os.listdir(fldr_path):
            if file.endswith('.ckpt'):
                ckpt_dict['ckpt_path'].append(os.path.join(fldr_path, file))
            if file.endswith('.csv'):
                ckpt_dict['inference_path'].append(os.path.join(fldr_path, file))

# ckpt_df = pd.DataFrame(ckpt_dict)
# ckpt_df
```

## **Aggregated Scores**

**Mask-RCNN ResNET 50 + FPN**

```python
run_score = pd.read_csv(os.path.join(root, 'metrics.csv'))
run_score.columns
```

```python
run_score['train/loss_step'].dropna().rolling(window=20, center=True).mean().plot(label='Train Loss Step')
run_score['train/loss_epoch'].dropna().plot(label='Train Loss Epoch')
plt.legend()
```

```python
run_score['val/mAP'].dropna().plot(label='Validation mAP')
plt.legend()
```

```python
run_score['lr-Adam'].dropna().plot(label='Adam Learning Rate')
plt.legend()
```
